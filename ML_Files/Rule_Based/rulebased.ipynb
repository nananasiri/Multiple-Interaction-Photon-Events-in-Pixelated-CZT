```python
import pandas as pd
df = {'ID':[], 'x':[], 'y':[], 'z':[] ,'T':[], 'E':[] }

with open("eShrinkBigBinnedBlurredFinal.csv", 'r') as f:
    counter = 0
    for line in f:
        sline = line.split('\t')
        if len(sline) == 25:
            df['ID'].append(int(sline[1]))
            df['x'].append(float(sline[13]))
            df['y'].append(float(sline[14]))
            df['z'].append(float(sline[15]))
            df['T'].append((float(sline[10])*1000000))
            df['E'].append(float(sline[11]))
        



#     print('number of NaN:', counter)
                
df = pd.DataFrame(df)
df.to_csv('energy.csv', index=False)
# df2 = pd.read_csv('ShrinkBigBinnedNoBlurredFinal6.csv')
df[0:20]

```

```python
inc_id_c = 0
dt_c = 0
X = []
y = []
df_gt = pd.read_csv('groundtruth.csv')

for i in range(0, len(df)-1, 3): 
    ID_list = [df['ID'].iloc[i], df['ID'].iloc[i+1], df['ID'].iloc[i+2]]
    row_list = [df.iloc[i], df.iloc[i+1], df.iloc[i+2]]
    
    row1 = df.loc[i, 'E':'E'].values.tolist() #add x, y, z and dt
    row2 = df.loc[i+1, 'E':'E'].values.tolist()
    row3 = df.loc[i+2, 'E':'E'].values.tolist()
    row1.extend(row2)
    row1.extend(row3)
    
    ID = df.loc[i,'ID']    
    gt_temp_rows = df_gt[df_gt['ID'] == ID] # find corresponding rows in grund truth dataframe
#     print(gt_temp_rows)
    
    if len(gt_temp_rows) != 3:
        continue
        
    idx = gt_temp_rows.index.tolist()[0] # read the first row's index
    if (gt_temp_rows.loc[idx, 'E':'E'].item() >= gt_temp_rows.loc[idx+1, 'E':'E'].item()) and (gt_temp_rows.loc[idx+1, 'E':'E'].item() >= gt_temp_rows.loc[idx+2, 'E':'E'].item()):
        label = 1
    else:
        label = 0
    
    X.append(row1)
    y.append(label)

X = np.array(X)
y = np.array(y)
```

```python
correct_pred = 0
for row, label in zip(X,y):
    if row[0] >= row[1] >= row[2]:
        pred = 1
    else:
        pred = 0
    
    if pred == label:
        correct_pred += 1
    
accuracy = correct_pred/len(X) * 100
print('Rule-based accuracy: ', accuracy, '%')
```
